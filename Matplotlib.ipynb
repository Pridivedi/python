```python
categories = ['A','B','C','D','E']
values = [5,7,3,8,4]
#plt.bar(x,y,color)
plt.bar(categories,values,color='blue',edgecolor='black')

#labels and title
plt.xlabel('student')# label for x axis
plt.ylabel('stats')#label for y axis
plt.title('mid term marks')#title of the graph

plt.grid(axis='y', linestyle ='--',alpha = 0.7)
#edge color='black'
#display the graph
plt.show()
```

```python
x = [0,1,2,3,4,5]
y = [0,1,4,9,16,25]
sizes=[35,200,60,130,100,80]
plt.scatter(x,y,c= 'red',s=sizes,alpha = 0.7,edgecolor='black',label='data points')

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('line plot example')

plt.legend()

plt.grid('true',linestyle='--',alpha=0.7)
```

```python
import matplotlib.pyplot as plt
import numpy as np
w= 0.2
Categories = ['Category A', 'Category B', 'Category C', 'Category D']  
Boys =[20, 35, 30, 35]  
Girls =[25, 32, 34, 20]
Others =[15, 27, 29, 31]  

bar1= np.arange(len(Categories))
plt.bar(bar1,Boys,w,color="lightcoral", label="Boys")
bar2 = [i+w for i in bar1]
plt.bar(bar2,Girls,w,color = "tan", label="Girls")

bar3 = [i+w for i in bar2]
plt.bar(bar3,Others,w,color = "aquamarine",label="Others")
plt.xticks(bar1+w/2,Categories)
plt.show()
```
